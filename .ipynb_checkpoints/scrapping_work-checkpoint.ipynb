# World War - I 

Primary purpose of this is to extract countries flags in the dataframe through the web-scrapping.

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://en.m.wikipedia.org/wiki/World_War_I#cite_note-1"
response = requests.get(url)
print(response)
data = BeautifulSoup(response.content, "html.parser")
```

```python
a = data.find('div', attrs={'class':'thumbinner'}).img['src']
```

```python
a
```

```python
from PIL import Image
import requests
from io import BytesIO

res = requests.get('https:'+a)
img = Image.open(BytesIO(res.content))
img
```

```python
data.div.table.get_text()
```

```python
data.select(".flagicon")[0].img['src'].split('/')[-2]
data.select(".flagicon")[1].img['src'].split('/')[-2]
```

```python
countries = []
for i in data.select(".flagicon"):
    a = i.img['src'].split('/')[-2]
    countries.append(a)
```

```python
flagicon = []
for i in data.select(".flagicon"):
    a = i.img['src']
    b = 'https:'+ a
    flagicon.append(b)
```

```python
flagicon
```

```python
countries
```

```python
import regex as re
country = []
for i in countries:
    a = re.findall("%=?\w+", i)
    b = "".join(a)
    c = i.replace(b,"")
    d = c.replace("_", " ")
    e = d.replace(".svg", "")
    f = re.findall("%\d+-\d+%\d+", e)
    g = "".join(f)
    h = e.replace(g, "")
    country.append(h)
```

```python
country
```

```python
country[2].replace('Flag of ', "")
```

```python
name = []
for i in country:
    a = i.replace("Flag of ", "")
    a = a.replace('the ', "")
    a = a.replace('Red Ensign','')
    a = a.replace('of ', '')
    name.append(a)
```

```python
name
```

```python
df = pd.DataFrame({'Name': name, 'Flag': flagicon})
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
def image_path(path):
    return '<img src="' + path + '"width ="50">'
from IPython.core.display import HTML
HTML(df.to_html(escape=False,formatters=dict(Flag=image_path)))
```
